**Learning curves**  

```python
import matplotlib.pyplot as plt
import numpy as np
x  = np.linspace(0.1,1,1000)
plt.plot(x, 10+1/x, label = r'$J_{cv}$')
plt.plot(x, 10+-1/x, label = r'$J_{train}$')
plt.xlabel(r'$m_{train}$ (training set size)')
plt.ylabel('error')
plt.legend()
plt.show()
plt.close()
```

 - Easy to fit to small amount of data
 - So training error low for small data set size
 - Adding data makes fitting to training examples more difficult
 - Which is why training error increases for data size
 - Cv always higher than J_train because you fit the model to training data

```python
import matplotlib.pyplot as plt
import numpy as np
x  = np.linspace(0.1,1,1000)
plt.title('High Bias')
plt.plot(x,10+1/x, label = r'$J_{cv}$')
plt.plot(x,10+-1/x, label = r'$J_{train}$')
plt.xlabel(r'$m_{train}$ (training set size)')
plt.hlines(y = 3, xmin = 0, xmax = 1, linestyle = '--', label = 'baseline')
plt.ylabel('error')
plt.legend()
plt.show()
plt.close()
```

If learning algo has high bias, getting more data will not help

```python
import matplotlib.pyplot as plt
import numpy as np
x  = np.linspace(0.1,0.3,1000)
plt.title('High Variance')
plt.plot(x,10+1/x, label = r'$J_{cv}$')
plt.plot(x,10+-1/x, label = r'$J_{train}$')
plt.xlabel(r'$m_{train}$ (training set size)')
plt.hlines(y = 10, xmin = 0, xmax = 1, linestyle = '--', label = 'baseline')
plt.ylabel('error')
plt.legend()
plt.show()
plt.close()
```

J_cv much higher than j_train  
Adding data will help  

Computationally expensive to make these learning curves so not done too much in practice
