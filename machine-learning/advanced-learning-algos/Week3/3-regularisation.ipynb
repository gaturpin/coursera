**Linear regression with regularisation**  

Large lambda:
- wieghts will be kept small
- close to 0
- so model will be f(x) ~ b; constant
- high bias - underfit

lambda = 0
- high variance - overfit
- small J_train, large J_cv

Choosing regularisation parameter:  
- Try different lambda values
- Calc J_cv for each model
- e.g. double lambda to some value
- Report test error of lowest J_cv model

**Bias and variance as a func of regularisation parameter, $\lambda$**  

something like this:

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,1,1000)
plt.plot(x, x**2, label = 'J_train')
plt.plot(x, 3*(x-0.6)**2+0.8, label = 'J_cv')
plt.xlabel('lambda')
plt.ylabel('error')
plt.legend()
```

Pick low J_cv value

**Establishing a baseline level of performance**  

Running example: speech recognition  
 - Say j_train = 10.8%
 - J_cv = 14.8%

 - J_train seems high so might say high bias, but what is human level performance  
 - Say human error is 10.6%  
 - So error not much higher than human performance  

 - Bigger J_cv is the problem here
 - Has a variance problem rather than bias.#

Important to know what is basline performance, e.g:  
 - baseline = 10.6%, J_train = 10.8%; J_cv = 14.8% - High variance because J_cv high compared to J_train
 - baseline = 10.6%, J_train = 15.0%, J_cv = 15.5% - High bais due to difference between baseline and J_train
 - Baseline = 10.6%, J_train = 15.0%, J_cv = 19.7% - High bais and high variance
