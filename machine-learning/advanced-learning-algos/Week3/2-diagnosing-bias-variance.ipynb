**Bias/variance**  

Indications of high bias/high variance
 - J_train high, J_cv high - high bias 
 - J_train low, J_cv high  - high variance
 - J_train low, J_cv low   - good fit

**Understanding bias and variance** 

Degree of polynomial d vs J:  
 - J_train: goes down from left to right, fits training data better and better
 - J_cv: U type curve with minimum - minimum is best order poly

High bias: J_train ~ J_cv  
High variance: J_cv >> J_train  
HIgh bias and high variance: J_train will be high and J_cv >> J_train  

Typically only have high bias or high variance, but can be both at the same time
