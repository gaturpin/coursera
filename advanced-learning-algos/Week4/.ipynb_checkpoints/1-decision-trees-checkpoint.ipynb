**Cat classification example**  

Tell me if an animal is a cat or not  


Features
 - Ear shape (floppy/pointy)
 - Face shape (round/not round)
 - Wiskers (present/not present)

Tree root top  
Leaf nodes bottom



**Learning process**  

 - Need to figure out what feature to split on at each node
 - Maximise purity / minimise impurity
 - Want to split into cats/not cats as best as possible

 - When to stop splitting?
 - When a node is 100% one class
 - When splitting a node results in tree exceeding maximum depth
   - Root is depth 0, layer down is depth 1 etc.
   - Limit depth of decision tree to make sure it doesn't get too big/less prone to overfitting
 - When improvements in purty are below a threshold
 - When number of examples in a node is below a threshold

**Measuring Purity**  

p_1 = fraction of examples that are cats  
Entropy as a measure of impurity:  
 - p_1 = 3/6, H(p_1) = 1

p0 = 1-p1  
H(p1) = -p1*np.log2(p1) - p0*np.log2(p0)  

Looks like:


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.001,0.999,1000)
p1 = x
p0 = 1-p1
y = -p1*np.log2(p1) - p0*np.log2(p0)
plt.plot(x, y)
plt.xlabel('p1')
plt.ylabel('H')
```

Note that 0log2(0) is undefined but we say it is 0 here

**Choosing a split**  

 - calculate information gain
 - Reduction in entropy from making a split:

information gain = H(initial data) - weighted average of split  

$ = H(p_1^{root}) - (w^{left}H(p_1^{left}) + w^{right} H(p_1^{right}))$

**Putting it all together**  

 - Start with all examples at root node
 - Calc information gain for all possible features
 - pick one with highest inf gain
 - Split dataset according to selected feature
 - Create left and right branches of tree
 - repeat splitting process unitl stopping criteria is met:
    - node is 100% one class
    - splitting a node exceeds maximum depth
    - info gain from more splits is less than a threshold
    - number of examples in node is below a threshold
